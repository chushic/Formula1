```python
import requests
import pandas as pd

response = requests.get('http://ergast.com/api/f1/current/driverStandings.json')

# Check if the request was successful
if response.status_code == 200:
    data = response.json()
    
    # Extract the relevant part of the JSON response
    driver_standings = data['MRData']['StandingsTable']['StandingsLists'][0]['DriverStandings']

    # Load into a DataFrame
    df = pd.json_normalize(driver_standings)
    print(df)
else:
    print(f"Failed to get data. HTTP Status Code: {response.status_code}")
```
